## 데이터 불러오기

```python
df=pd.read_csv("data.csv")
df
```

### 팀별 데이터 프레임 생성

```python
기획=df[df['소속팀']=="기획"]
개발=df[df['소속팀']=="개발"]
디자인=df[df['소속팀']=="디자인"]
기획.reset_index(inplace=True)
기획.drop(['index'], inplace=True, axis=1)
개발.reset_index(inplace=True)
개발.drop(['index'], inplace=True, axis=1)
디자인.reset_index(inplace=True)
디자인.drop(['index'], inplace=True, axis=1)
```

```python
요일df_디자인=디자인['조모임 가능 요일'].str.split('/').tolist()
요일df_디자인
```

### 요일별로 데이터프레임 생성

```python
요일list=['월요일', '화요일', '수요일', '목요일', '금요일', '일요일오전', '일요일오후', '일요일저녁']
요일df=pd.DataFrame([[0]*len(요일list)]*len(디자인), columns=요일list)
요일df['이름']=""
요일df['디자인 역량 점수']=0
for i in range(len(요일df)):
    요일df.loc[i, '이름']=디자인.loc[i, '이름']
    요일df.loc[i, '디자인 역량 점수']=디자인.loc[i, '디자인 역량 점수']
# 요일df['이름']=디자인['이름']
# 요일df['디자인 역량 점수']=디자인['디자인 역량 점수']
요일df
```

### 가능 요일에 1 표시

```python
for i in range(len(디자인)):
    가능요일=디자인.loc[i, '조모임 가능 요일'].split('/')
    for j in range(len(요일list)):
        if 요일list[j] in 가능요일:
            요일df.loc[i, 요일list[j]]+=1
요일df
```

### 가능한 요일 개수 count

```python
요일df['요일개수']=0
for i in range(len(요일df)):
    total=0
    for j in range(0, 8):
        total+=요일df.iloc[i,j]
    요일df.loc[i,'요일개수']=total
요일df  
```

### 가능한 요일 개수 적은 순, 디자인 역량점수 높은 순으로 정렬

```python
요일df.sort_values(by=['요일개수', '디자인 역량 점수'], inplace=True, ascending=[True, False])
요일df.reset_index(inplace=True)
요일df.drop(['index'], inplace=True, axis=1)
요일df
```

#### (요일, 가능인원총합)으로 배열 생성

```python
가능요일list=[]# 요일, 총합
for i in range(len(요일list)):
    가능요일list.append([i, sum(요일df.iloc[:, i])])
가능요일list.sort(key=lambda x:x[1])
가능요일list
```

### 가능인원이 적은 요일 순서대로 정렬

```python
디자인_팀= [[], [],[], [], [], [],[], [], []]
디자인팀원=list(디자인['이름'])
for i in range(len(가능요일list)):
    did=0
    for j in range(len(요일df)):  
        if did==0 and 요일df.iloc[j, 가능요일list[i][0]]==1 and 요일df.loc[j, '이름'] in 디자인팀원:
            did=1
            디자인_팀[가능요일list[i][0]].append(요일df.loc[j, '이름'])
            디자인팀원.remove(요일df.loc[j, '이름'])
            print(디자인_팀)
디자인_팀[len(가능요일list)]=[디자인팀원[0]]
```
